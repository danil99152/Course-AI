```python
import numpy as np
```

Функция активации $sig(x)=\frac{1}{1+exp^{-x}}$:

```python
def sigmoid(x): 
     return 1/(1+np.exp(-x))
```

Частная производная:

```python
def deriv(x): 
     return x*(1-x)
```

3 Нейрона на ввод:

```python
A = np.array([0,0,1,0,0]) #На улице дождь? -0.7
B = np.array([0,0,1,1,0]) #Есть ли джакузи? 1
C = np.array([0,1,0,1,0]) #Преобладает ли там противоположный пол? 0.5

mnim = np.array([-0.7,1,0.5])

X = np.array([A,B,C]).T
X.shape
```

1 Нейрон на вывод:

```python
Y = np.array([[0,1,0,1,0]]).T
Y.shape
```

Случайные веса для каждого нейрона входа:

```python
weights = np.random.random((3,1))
print("Случайные веса:",weights)
```

```python
def train(X,Y,weights,alpha,epochs): #обучение
    for epoch in range(epochs):
        outputs = sigmoid(np.dot(X,weights)) #суммирование в выходной нейрон
        abs_err = ((outputs - Y)**2).sum()/ (2*len(Y)) #Квадратичная ошибка для нашего понимания
        adj = np.dot(X.T,alpha*abs_err*deriv(outputs))#На сколько надо изменить веса
        print(f"{epoch} epoch: {abs_err}")
        weights += adj#Уменьшаем веса

    return weights
```

```python
weights = train(X,Y,weights,0.01,1000)
```
